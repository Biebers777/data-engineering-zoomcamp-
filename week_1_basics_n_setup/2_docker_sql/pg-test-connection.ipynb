````python
df = pd.read_csv('../yellow_tripdata_2021-01.csv')
````

````python
zones = pd.read_csv('../taxi+_zone_lookup.csv')
````

````python
df
````

````python
df.tpep_pickup_datetime = pd.to_datetime(df.tpep_pickup_datetime)
df.tpep_dropoff_datetime = pd.to_datetime(df.tpep_dropoff_datetime)
````

````python
df['tpep_pickup_date'] = df.tpep_pickup_datetime.dt.date
````

````python
df
````

````python
df.groupby("tpep_pickup_date").max().sort_values('tip_amount').tail(5)
````

````python
jan14 = df[df.tpep_pickup_date == datetime.date(2021, 1, 14)]\
.groupby("DOLocationID").count().sort_values('tip_amount').tail(5)
````

````python
jan14[["VendorID"]].join(zones.set_index("LocationID"))
````

````python
zones
````

````python
cand = df.groupby(['PULocationID', 'DOLocationID']).mean()[['total_amount']].sort_values('total_amount').tail(1)
cand = cand.reset_index()
````

````python
pd.merge(cand, zones, left_on="PULocationID", right_on="LocationID")
````

````python
pd.merge(cand, zones, left_on="DOLocationID", right_on="LocationID")
````

````python
df.to_sql(name='yellow_tripdata_trip', con=engine, index=False)
````

````python
query = """
SELECT * FROM yellow_tripdata_trip LIMIT 10
"""

pd.read_sql(query, con=engine)
````
