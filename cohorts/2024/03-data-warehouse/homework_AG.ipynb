# CREATING COMBINED PARQUET FILE: 2022

````python
import pandas as pd
````

````python
dfs = []

for idx in range(12):

    df = pd.read_parquet(f'green_tripdata_2022-{idx + 1:02d}.parquet')
    dfs.append(df)

result_df = pd.concat(dfs, ignore_index=True)    
````

````python
result_df
````

````python
result_df.to_parquet('./green_tripdata_2022.parquet')
````
